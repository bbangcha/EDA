## 4. 두데이터 합치기

```python
left = pd.DataFrame({
    "key": ["K0", "K4", "K2", "K3"],
    "A": ["A0", "A1", "A2", "A3"],
    "B": ["B0", "B1","B2", "B3"]
})
left 
# 딕셔너리 안의 리스트 형태의 데이터프레임
```

```python
right = pd.DataFrame([
    {"key":"K0", "C":"C0", "D":"D0"},
    {"key":"K1", "C":"C1", "D":"D1"},
    {"key":"K2", "C":"C2", "D":"D2"},
    {"key":"K3", "C":"C3", "D":"D3"},
])
right
# 리스트 안의 딕셔너리 형태의 데이터프레임
```

#### how = inner가 디폴트되어 생략된 부분으로 공통된 부분만 출력 

```python
pd.merge(left, right,  on="key")
```

#### outer는 합집합으로 모든 데이터 출력

```python
# .merge(df_name, df_name, how = 데이터프레임 조건, on = 병합기준 컬럼)

pd.merge(left, right, how = 'outer',  on="key")
```

#### 오른쪽을 기준으로 데이터가 작성되어야 하므로 A, B 컬럼에 없는 K1 인덱스 value는 NaN으로 표시

```python
pd.merge(left, right, how="right", on="key")
```

#### 왼쪽을 기준으로 데이터가 작성되어야 하므로 C, D 컬럼에 없는 K4 인덱스 value는 NaN으로 표시

```python
pd.merge(left, right, how="left", on="key")
```

#### Numpy를 이용한 1차 직선 만들기 
- np.polyfit(): 직선을 구성하기 위한 계수를 계산
- np.poly1d(): polyfit 으로 찾은 계수로 파이썬에서 사용할 수 있는 함수로 만들어주는 기능 

```python
import numpy as np 

fp1 = np.polyfit(data_result["인구수"], data_result["소계"], 1)
# np.polyfit : 입력한 x, y 값에 대한 함수를 만들기 위한 기울기와 y절편 계수 계산
# data_result["인구수"] : x값, data_result["소계"] : y값, 1 : n차 방정식 설정
fp1
# array = 1.11155868e-03 : 기울기, 1.06515745e+03 : y절편
```

```python
f1 = np.poly1d(fp1)
# np.polyfit에서 찾은 계수로 파이썬에서 사용할 함수를 만들어 줌
f1
```

```python
f1(400000)
# f1 (= np.poly1d) 은 함수로서 x값 입력 시 y값 출력
```

- 인구가 40만인 구에서 서울시의 전체 경향에 맞는 적당한 CCTV 수는? 

```python
fx = np.linspace(100000, 700000, 100)
# 경향선을 그리기 위한 X 데이터 생성 
# np.linspace(a, b, n): a부터 b까지 n개의 등간격 데이터 생성 
```

```python
def drawGraph():
    
    plt.figure(figsize=(5, 5))
    plt.scatter(data_result["인구수"], data_result["소계"], s=20)
    # x값 : data_result["인구수"], y값 : data_result["소계"], s : 점의 크기
    plt.plot(fx, f1(fx), ls="dashed", lw=3, color="g")
    plt.xlabel("인구수")
    plt.ylabel("CCTV")
    plt.grid(True)
    plt.show() 
drawGraph()
```

## 7. 강조하고 싶은 데이터를 시각화해보자

### 그래프 다듬기 
#### 경향과의 오차 만들기 
- 경향(trend)과의 오차를 만들자 
- 경향은 f1 함수에 해당 인구를 입력 
- f1(data_result["인구수"])

```python
fp1 = np.polyfit(data_result["인구수"], data_result["소계"], 1) 
f1 = np.poly1d(fp1) 
fx = np.linspace(100000, 700000, 100) 
```

```python
data_result.head(3)
```

```python
data_result["오차"] = data_result["소계"] - f1(data_result["인구수"])
```

```python
data_result.head(1)
```

```python
# 경향과 비교해서 데이터의 오차가 너무 나는 데이터를 계산 

df_sort_f = data_result.sort_values(by="오차", ascending=False) # 내림차순 
df_sort_t = data_result.sort_values(by="오차", ascending=True) # 오름차순 
```

```python
# 경향 대비 CCTV를 많이 가진 구 
df_sort_f.head()
```

```python
# 경향 대비 CCTV를 적게 가진 구 
df_sort_t.head()
```

```python
from matplotlib.colors import ListedColormap

# colormap 을 사용자 정의(user define)로 세팅 
color_step = ["#e74c3c", "#2ecc71", "#95a9a6", "#2ecc71", "#3498db", "#3498db"]
my_cmap = ListedColormap(color_step)
```

```python
def drawGraph():
    
    plt.figure(figsize=(5, 5))
    plt.scatter(data_result["인구수"], data_result["소계"], s=20, c=data_result["오차"], cmap=my_cmap)
    # x값 : data_result["인구수"], y값 : data_result["소계"], s : 점의 크기, c : color 세팅 기준
    plt.plot(fx, f1(fx), ls="dashed", lw=2, color="g")
    # x값 : fx, y값 : f1(fx), ls : line stlye, lw : line weight color : line color 

    for n in range(5):
        # 상위 5개 
        plt.text(
            df_sort_f["인구수"][n] * 1.02, # x 좌표
            df_sort_f["소계"][n] * 0.98,  # y 좌표
            df_sort_f.index[n], # title 
            fontsize=10,
        )
    
        # 하위 5개 
        plt.text(
            df_sort_t["인구수"][n] * 1.02, 
            df_sort_t["소계"][n] * 0.98,
            df_sort_t.index[n],
            fontsize=10
        )
        
    
    plt.xlabel("인구수")
    plt.ylabel("CCTV")
    plt.colorbar()
    plt.grid(True)
    plt.show() 
drawGraph()
```
